```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data (assuming 'df' already exists)
x1 = pd.to_numeric(df["O3_DU"], errors='coerce')
x2 = pd.to_numeric(df["Temp"], errors='coerce')  
y = pd.to_numeric(df["Press"], errors='coerce')

# Create the plot
fig, ax = plt.subplots(figsize=(20, 15))

# Plot both lines
ax.plot(x1, y, label="O3_DU")
ax.plot(x2, y, label="Temp")

# Set x-axis ticks for O3_DU (optional customization)
ax.set_xticks(np.arange(-90, 290, 20))

# Invert the y-axis (higher values at the top)
ax.invert_yaxis()

# Add labels and title
#ax.set_xlabel("DU Value")
ax.set_ylabel("Pressure")
ax.set_title("O3_DU and Temp U vs Pressure")

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure data is numeric
x_o3 = pd.to_numeric(df["O3_mPa"], errors='coerce')
x_temp = pd.to_numeric(df["Temp"], errors='coerce')  
y_alt = pd.to_numeric(df["GeopAlt"], errors='coerce')

# Drop rows with NaNs to avoid plotting issues
mask = ~(x_o3.isna() | x_temp.isna() | y_alt.isna())
x_o3 = x_o3[mask]
x_temp = x_temp[mask]
y_alt = y_alt[mask]

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 8))

# Plot O3_mPa vs GeopAlt on the main x-axis
ax1.plot(x_o3, y_alt, color='tab:blue', label='O3_mPa')
ax1.set_xlabel("O3 (mPa)", color='tab:blue')
ax1.tick_params(axis='x', labelcolor='tab:blue')

# Create a twin x-axis for Temp
ax2 = ax1.twiny()
ax2.plot(x_temp, y_alt, color='tab:red', label='Temp')
ax2.set_xlabel("Temperature (°C or K)", color='tab:red')
ax2.tick_params(axis='x', labelcolor='tab:red')

# Invert the y-axis to represent increasing altitude
#ax1.set_ylabel("Geopotential Altitude (m)")
#ax1.invert_yaxis()

# Title and layout
plt.title("Vertical Profile of O3_mPa and Temperature vs GeopAlt")
plt.grid(True)
plt.tight_layout()
plt.show()
```
